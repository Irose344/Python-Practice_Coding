```python
'''2. Write a Python program to solve quadratic equation.
The standard form of a quadratic equation is:
where
a, b and c are real numbers and
The solutions of this quadratic equation is given by:
𝑎𝑥 + 𝑏𝑥 + 𝑐 = 0
2
𝑎 ≠ 0
(−𝑏 ± (𝑏 − 4𝑎𝑐^)1/2)/(2𝑎))
'''
import math

#input a,b,c values from the user

a=int(input("Enter a:"))
b=int(input("Enter b:"))
c=int(input("Enter c:"))


#find out the discriminant-based on the value of discriminant, we can find out how the root values will be

dis=(b**2)-(4*a*c)

#if discriminant>0, two real and distinct roots
if dis>0:
  r1= (-b * math.sqrt(dis))/(2*a)
  r2= (-b * math.sqrt(dis))/(2*a)
  print(f"roots are {r1},{r2}")
#if discriminant=0, only 1 real root
elif dis==0:
  r1=-b/(2*a)
  print(f"root: {r1}")
#if discriminant<0, there are 2 complex roots
else:
  real=-b/(2*a)
  imaginary=math.sqrt(abs(dis))/(2*a)
  print(f"r1: {real}-{imaginary}i")
  print(f"r1: {real}+{imaginary}i")



```
